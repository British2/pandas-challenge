```python
#File to load
file_to_load = "C:\\Data\\purchase_data.csv"

#Read Purchase File and Store in Pandas
purchase_data = pd.read_csv(file_to_load)
df = pd.DataFrame(purchase_data)
df.head()
```

```python
# Player Count

players_df = purchase_data.loc[:, ["SN", "Age", "Gender"]]
players_df = players_df.drop_duplicates()
num_of_players = players_df.count()[0]
pd.DataFrame({"Total Players": [num_of_players]})
```

```python
# Purchase Analysis (Total)

number_of_unique_items = len(df['Item ID'].unique())
AP = round(float(df['Price'].mean()),2)
num_of_purchases = len(df['Price'])
total_rev = float(df['Price'].sum())


Analysis_df = pd.DataFrame({
    'Number of Unique Items': [number_of_unique_items],
    'Average Price': '$' + str(AP),
    'Number of Purchases': [num_of_purchases],
    'Total Revenue': '$' + str(total_rev)
})

Analysis_df
```

```python
df_gender = df[['Gender','SN']].drop_duplicates(subset = 'SN')

gender_num = df_gender['Gender'].value_counts(0)
gender_percent = df_gender['Gender'].value_counts(1)

gender_num_df = pd.DataFrame(gender_num)
gender_percent_df = round(pd.DataFrame(gender_percent) * 100, 2)

gender_summary_df = gender_num_df.merge(gender_percent_df, left_index = True, right_index = True)
gender_summary_df.columns = ['Total Count', 'Percentage of Players']
gender_summary_df["Percentage of Players"]= gender_summary_df["Percentage of Players"].map("{:,.2f}%".format)
gender_summary_df
```

```python
# Purchase Analysis

purchase_by_gender_df = df.groupby('Gender')

purchase_count = round(purchase_by_gender_df['Purchase ID'].count(), 0)
avg_purchase_price = round(purchase_by_gender_df['Price'].mean(), 2)
total_purchase_value = round(purchase_by_gender_df['Price'].sum(), 2)
purchase_value_per_gender = round(total_purchase_value / gender_num, 2)

Gender_Analysis_df = pd.DataFrame({
    'Purchase Count': purchase_count,
    'Average Purchase Price': avg_purchase_price,
    'Total Purchase Value': total_purchase_value,
    'Avg Total Purchase per Person': purchase_value_per_gender
})

Gender_Analysis_df

Gender_Analysis_df['Average Purchase Price']= Gender_Analysis_df['Average Purchase Price'].map("${:,.2f}".format)
Gender_Analysis_df

Gender_Analysis_df['Total Purchase Value']= Gender_Analysis_df['Total Purchase Value'].map("${:,.2f}".format)
Gender_Analysis_df

Gender_Analysis_df['Avg Total Purchase per Person']= Gender_Analysis_df['Avg Total Purchase per Person'].map("${:,.2f}".format)
Gender_Analysis_df
```

```python
# Age Demographic

bins = [0, 9, 14, 19, 24, 29, 34, 39, 999]
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

df_gender = df[['Age','SN']].drop_duplicates(subset = 'SN')

players_df["Age Groups"] = pd.cut(players_df["Age"], bins, labels = bin_labels)

age_totals = players_df["Age Groups"].value_counts(0)
age_percents = age_totals / num_of_players * 100

age_dem = pd.DataFrame({"Total Purchase Count": age_totals, "Percentage of Players": age_percents}) 

age_dem['Percentage of Players']= age_dem['Percentage of Players'].map("{:,.2f}%".format)

age_dem.sort_index()
```

```python
# Purchase Analysis (Age)

df["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=bin_labels)

a_count = purchase_data.groupby(["Age Ranges"]).count()["Price"].rename("Purchase Count")
a_avg = purchase_data.groupby(["Age Ranges"]).mean()["Price"].rename("Average Purchase Price")
a_total = purchase_data.groupby(["Age Ranges"]).sum()["Price"].rename("Total Purchase Value")
a_demo_avg = a_total / age_dem["Total Purchase Count"]

a_data = pd.DataFrame({"Purchase Count": a_count, "Average Purchase Price": a_avg, "Total Purchase Value": a_total, "Avg Total Purchase per Person": a_demo_avg})

a_data ["Purchase Count"] = a_data["Purchase Count"].map("{:,}".format)
a_data["Average Purchase Price"] = a_data["Average Purchase Price"].map("${:,.2f}".format)
a_data["Total Purchase Value"] = a_data["Total Purchase Value"].map("${:,.2f}".format)
a_data["Avg Total Purchase per Person"] = a_data["Avg Total Purchase per Person"].map("${:,.2f}".format)
a_data = a_data.loc[:, ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"]]


a_data.reset_index(inplace=True)
a_data.rename(columns={"index": "Age Ranges"}, inplace=True)

a_data
```

```python
# Top Spenders

users = df.groupby(["SN"]).count()["Price"].rename("Purchase Count")
user_avg = df.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
user_total = df.groupby(["SN"]).sum()["Price"].rename("Total Purchase Value")

user_data = pd.DataFrame({"Purchase Count": users, "Average Purchase Price": user_avg, "Total Purchase Value": user_total})

user_sorted = user_data.sort_values("Total Purchase Value", ascending = False)

user_sorted["Average Purchase Price"] = user_sorted["Average Purchase Price"].map("${:,.2f}".format)
user_sorted["Total Purchase Value"] = user_sorted["Total Purchase Value"].map("${:,.2f}".format)
user_sorted = user_sorted.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

user_sorted.head(5)
```

```python
# Most Popular Item

mp_items = df.loc[:,["Item ID", "Item Name", "Price"]]

purchase_count = mp_items.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
price = mp_items.groupby(["Item ID", "Item Name"]).mean()["Price"]
total_purchase_value = mp_items.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Value")

mp_item_df = pd.DataFrame({"Purchase Count": purchase_count, "Item Price": price, "Total Purchase Value": total_purchase_value})

items_sorted = mp_item_df.sort_values("Purchase Count", ascending = False)

items_sorted["Purchase Count"] = items_sorted["Purchase Count"].map("{:,}".format)
items_sorted["Item Price"] = items_sorted["Item Price"].map("${:,.2f}".format)
items_sorted["Total Purchase Value"] = items_sorted["Total Purchase Value"].map("${:,.2f}".format)
most_liked_items = items_sorted.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]

most_liked_items.head(5)
```

```python
# Most Profitable Items

total_value = mp_item_df.sort_values("Total Purchase Value", ascending = False)

total_value["Purchase Count"] = total_value["Purchase Count"].map("{:,}".format)
total_value["Item Price"] = total_value["Item Price"].map("${:,.2f}".format)
total_value["Total Purchase Value"] = total_value["Total Purchase Value"].map("${:,.2f}".format)
biggest_profit = total_value.loc[:,["Purchase Count", "Item Price", "Total Purchase Value"]]

biggest_profit.head(5)
```
